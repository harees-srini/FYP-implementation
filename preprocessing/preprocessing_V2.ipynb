```python
# Iterate over unique place IDs and get place types
unique_place_ids = rawdata['placeID'].unique()
for place_id in unique_place_ids:
    # Prepare request parameters
    params = {
        'place_id': place_id,
        'fields': 'types,name',
        'key': api_key
    }
    
    # Send request to Places API
    response = requests.get(endpoint, params=params)
    data = response.json()
    
    # Extract place types from response
    place_types = data.get('result', {}).get('types')
    place_name = data.get('result', {}).get('name')
    
    # Print place types
    if place_types and place_name:
        print(f"Place ID: {place_id}, Place Types: {place_types}, Name: {place_name}")
    else:
        print(f"No place types found for Place ID: {place_id}")
```

```python
def get_place_types(place_id):
    try:
        place = gmaps.place(place_id)
        if 'types' in place['result']:
            return place['result']['types']
        else:
            return None
    except Exception as e:
        print(f"Error fetching place types for Place ID {place_id}: {e}")
        return None
```

```python
# Add a new column for place types
rawdata['new_place_types'] = None

# Iterate through placeId column
for index, row in rawdata.iterrows():
    place_id = row['placeID']    
    place_types = get_place_types(place_id)
    if place_types:
        rawdata.at[index, 'new_place_types'] = ','.join(place_types)

rawdata.head(10)
```
